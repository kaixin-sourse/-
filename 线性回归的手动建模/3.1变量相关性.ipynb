```python
X = np.random.randn(20)
```

```python
y = X + 1
np.corrcoef(X,y)
```

```python
#对应折线图
plt.plot(X,y)
```

```python
plt.plot(X,y,'o')
```

```python
a = y.shape
a
```

```python
# 创建一个和y一样形状的俯冲标准正态分布的随机数组
np.random.normal(size = a)
```

```python
ran = np.random.normal(size = y.shape)
ran
```

```python
np.random.normal?
```

```python
#白噪声，扰动项
delta = 0.5
```

```python
r = ran * delta
r
```

```python
#在y上增加扰动项
y1 = y + r
y1
```

```python
#一张画布上的多子图
plt.subplot(121)
plt.plot(X,y,'o')
plt.title("y = x + 1")
plt.subplot(122)
plt.plot(X,y1,'o')
plt.title("y = x + 1 + r")
```

```python
# delta系数取值数组
dl = [0.5,0.7,1,1.5,2,5]
```

```python
# 空列表容器
yl = []
cl = []
```

```python
for i in dl:
    yn = X + 1 + (ran * i)
    cl.append(np.corrcoef(X,yn))
    yl.append(yn)
```

```python
cl
```

```python
#绘制y与x的相关性分部曲线
for i in yl:
    plt.plot(X,i,'o')
    plt.title("y = x + i + delta")
    
```

```python
for i in range(1,7):
    plt.subplot(230 + i)
    plt.plot(X,yl[i - 1],'o')
    plt.plot(X,y,'r-') #初始排列关系，相关性。红线
    
```
